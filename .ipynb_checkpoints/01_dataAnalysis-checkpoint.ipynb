```python
attrition = data[data['Attrition'] == 'Yes']
no_attrition = data[data['Attrition'] == 'No']

plt.figure(figsize=(18,5))
plt.suptitle('Attrition')
plt.subplot(121)
ax = sns.countplot(x='Attrition', data=data)
for i in ax.patches:
    ax.annotate("%.0f" % i.get_height(), (i.get_x() + i.get_width() / 2., i.get_height()),
                 ha="center", va="center", fontsize=11, rotation=0, xytext=(0, 5),
                 textcoords="offset points")

plt.subplot(122)
plt.pie(data['Attrition'].value_counts(), labels=['Yes', 'No'], autopct='%1.1f%%')
plt.title('Percentage')
plt.show()
```

```python
# Grouping age
def groupAge(data):
    data.loc[data['Age'] <= 20,'Age'] = 1
    data.loc[(data['Age'] > 20) & (data['Age'] <= 30), 'Age'] = 2
    data.loc[(data['Age'] > 30) & (data['Age'] <= 40), 'Age'] = 3
    data.loc[(data['Age'] > 40) & (data['Age'] <= 50), 'Age'] = 4
    data.loc[(data['Age'] > 50) & (data['Age'] <= 65), 'Age'] = 5

groupAge(data)
```

## Drop Unnecessary Features
- Drop `Over18` because every data is the same
- Drop `EmployeeCount` because every data is the same
- Drop `StandardHours` because every data is the same

```python
df = data.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1)
```

```python
df.columns
```

```python
df.head()
```

## Labeling Data

```python
# Grouping age
def groupAge(df):
    df.loc[df['DistanceFromHome'] <= 10,'DistanceFromHome'] = 1
    df.loc[(df['DistanceFromHome'] > 10) & (df['DistanceFromHome'] <= 20), 'DistanceFromHome'] = 2
    df.loc[(df['DistanceFromHome'] > 20) & (df['DistanceFromHome'] <= 30), 'DistanceFromHome'] = 3
    
groupAge(df)
```

```python
df[['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus']].head()
```

```python
dfDummy = pd.get_dummies(df[['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus']])
dfDummy
```

```python
label = LabelEncoder()
df['Attrition'] = label.fit_transform(df['Attrition'])
df['Gender'] = label.fit_transform(df['Gender'])
df['OverTime'] = label.fit_transform(df['OverTime'])
```

```python
df
```

## Drop Feature after Labeling

```python
df = df.drop(['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus'], axis=1)
```

```python
df = pd.concat([df, dfDummy], axis=1)
```
